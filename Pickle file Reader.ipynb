```python
import pickle

svm_loaded = pickle.load(open("svm_model.pkl", "rb"))

print(type(svm_loaded))
```

```python
vectorizer_loaded = pickle.load(open("tfidf_vectorizer.pkl", "rb"))
encoder_loaded = pickle.load(open("label_encoder.pkl", "rb"))
```

```python
sample_text = ["I feel very lonely and hopeless about my future"]

sample_tfidf = vectorizer_loaded.transform(sample_text)

pred = svm_loaded.predict(sample_tfidf)

print("Encoded Prediction:", pred[0])
print("Actual Label:", encoder_loaded.inverse_transform(pred)[0])
```
